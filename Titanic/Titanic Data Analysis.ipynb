## Data Wrangling

```python
titanic_data.isnull()
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.drop(["cabin","body",'home.dest','boat'], axis=1, inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.dropna(inplace= True)
```

```python
sns.heatmap(titanic_data.isnull(),yticklabels=False, cbar=False)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.head()
```

```python
sex = pd.get_dummies(titanic_data["sex"], drop_first=True)
sex.head()
```

```python
embark = pd.get_dummies(titanic_data["embarked"], drop_first=True)
embark.head()
```

```python
pcl = pd.get_dummies(titanic_data['pclass'],drop_first=True)
pcl.head()
```

```python
titanic_data = pd.concat([titanic_data,sex,embark,pcl], axis=1)
titanic_data.head()
```

```python
titanic_data.drop(["pclass","sex",'embarked','name','ticket'],axis=1, inplace=True)
```

```python
titanic_data.head()
```

## Train Data 

```python
X= titanic_data.drop('survived',axis=1) #indepedent variable
y = titanic_data['survived']            #dependent variable
pd.DataFrame(y).head(10)
# titanic_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
print(type(X_train))
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

## Accuracy Check

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
